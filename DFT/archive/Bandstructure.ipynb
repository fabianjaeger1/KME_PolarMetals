```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": ["Computer Modern Roman"]})


plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
plt.rcParams["figure.figsize"]=(8, 6)
# load data

Fermi_energy = 6.1605

data = np.loadtxt('Bandx.dat.gnu')
k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k))) - Fermi_energy
print(np.shape(bands))
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')
plt.xlim(min(k), max(k))
# Fermi energy


plt.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)


# High symmetry k-points (check bands_pp.out)

plt.axvline(0.0000, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.1281, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.2563, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.4375, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.5578, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.6859, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.8141, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")


# text labels


plt.xticks(ticks= [0, 0.1281, 0.2563, 0.4375, 0.5578, 0.6859, 0.8141], \
           labels=['$\Gamma$', 'X', 'M', '$\Gamma$', 'Z', 'R', 'A'])
plt.ylabel("Energy (eV)")
plt.ylim(-6, 6)
#plt.text(2.3, 5.6, 'Fermi energy', fontsize= 'small')
plt.savefig('PTO_bandstructure.pdf')
plt.show()
```
