```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": ["Computer Modern Roman"]})


plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
plt.rcParams["figure.figsize"]=(8, 6)
```

```python
path_dft_band = "Bandstructure_files"
path_df_band_wannier = "Bandstructure_wannier_files"
path_plots = "Plots"
```

# Bandstructure DFT

```python
Fermi_energy = 6.1605

data = np.loadtxt('{}/Bandx.dat.gnu'.format(path_dft_band))
k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k))) - Fermi_energy
print(np.shape(bands))
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth=1, alpha=0.5, color='k')
plt.xlim(min(k), max(k))

plt.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)

plt.axvline(0.0000, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.1281, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.2563, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.4375, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.5578, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.6859, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.8141, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")


plt.xticks(ticks= [0, 0.1281, 0.2563, 0.4375, 0.5578, 0.6859, 0.8141], \
           labels=['$\Gamma$', 'X', 'M', '$\Gamma$', 'Z', 'R', 'A'])
plt.ylabel("Energy (eV)")
plt.ylim(-6, 6)
plt.savefig('{}/PTO_bandstructure.pdf'.format(path_plots))
plt.show()
```

## Bandstructure Combined

```python
ticks_1 = [0, 0.1281, 0.2563, 0.4375, 0.5578, 0.6859, 0.8141]
ticks_2 = [0, 0.8051280111, 1.6102560221, 2.7488789748, 3.5048004135, 4.3099284246, 5.1150564357]
position = 2
scaling = ticks_2[position]/ticks_1[position]
print(scaling)
```

```python
Fermi_energy = 6.1605

data = np.loadtxt('{}/Bandx.dat.gnu'.format(path_dft_band))
k = np.unique(data[:, 0] * scaling)
bands = np.reshape(data[:, 1], (-1, len(k))) - Fermi_energy
print(np.shape(bands))
for band in range(len(bands)):
    plt.plot(k, bands[band, :], linewidth =0.75, alpha=1, color='k')
plt.xlim(min(k), max(k))


data_wannier = np.loadtxt("{}/pto_band.dat".format(path_df_band_wannier))
k_wannier = np.unique(data_wannier[:, 0])
bands_wannier = np.reshape(data_wannier[:, 1], (-1, len(k_wannier))) - Fermi_energy
for band in range(len(bands_wannier)):
    plt.plot(k_wannier, bands_wannier[band, :], linewidth=0.5, alpha=0.75, color='r', linestyle = '--')
plt.xlim(min(k_wannier), max(k_wannier))

plt.axhline(0, linestyle=(0, (5, 5)), linewidth=1, color='k', alpha=1)


plt.axvline(0.0000, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(0.8051280111, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(1.6102560221, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(2.7488789748, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(3.5048004135, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(4.3099284246, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")
plt.axvline(5.1150564357, linewidth=0.75, color='k', alpha=0.5,linestyle = "--")

plt.xticks(ticks= [0, 0.8051280111, 1.6102560221, 2.7488789748, 3.5048004135, 4.3099284246, 5.1150564357], 
            labels=['$\Gamma$', 'X', 'M', '$\Gamma$', 'Z', 'R', 'A'])
plt.ylabel("Energy (eV)")
plt.ylim(-6, 6)

plt.savefig('{}/PTO_bandstructure_combined.pdf'.format(path_plots))
plt.show()
```
